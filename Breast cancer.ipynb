## Making Use of Predictive Analysis to Forecast Breast Tumor Diagnosis

Chapter 1 : It is the most common malignant tumor in females globally, thus being among the leading contributors towards the total cases of cancer, excluding its placement as the leading cause of female mortality due to a tumor the world over. Thus, it is considered one of the most important public health concerns for which awareness, besides early detection, is crucial towards achieving better prognosis and minimizing mortality (World, 2024).

Cancer of the breast occurs when there is abnormal, uncontrolled growth of the cells of the breast, either in the form of a mass or lump, otherwise known as a tumor. This is where the statement that not all tumors are cancerous becomes relevant. In fact, tumors can be benign, pre-malignant, or malignant. Of these, the first are benign, which are non-cancerous and do not metastasize to other parts of the body and the second are termed as pre-malignant or pre-cancerous as they have the possibility of turning cancerous over a period of time, whereas malignant tumors are cancerous invade the surrounding tissues and metastasize to other parts of the body (Breastcancer.org, 2022).

It has been documented that the early detection of cancer in the breast will ensure effective treatment and increased survival chances. A host of diagnostic tools help in the identification and investigation of abnormalities in the breast. Among the more common one-worldwide screening methods, the mammogram is an X-ray picture of the breast that may disclose tumors not able to be felt because they are too small. MRI, or Magnetic Resonance Imaging, is more detailed in its imaging and is often used for high-risk patients or for examining any abnormalities found on a mammogram (Shetty, 2010).

Ultrasound, which takes pictures of breast tissue through the use of sound waves, is often used to distinguish between a solid mass and a fluid-filled cyst. If an imaging test reveals a suspicious abnormality, a biopsy will be performed. A biopsy involves removing a small section of the breast tissue, looking at it under a microscope to confirm if cancer cells are present, and if cancer is present, the type and extent of its aggressiveness(Clinic, 2023). These diagnostic tools have a major role in early detection of the disease and hence also provide more options for cure. Regular screening is all the more important in people presenting risk factors due to family history, genetic predisposition, or even certain aspects of lifestyle. Global awareness campaigns and educational efforts have encouraged women to seek out regular check-ups and be proactive regarding the health of their breasts. Thus, early and regular detection on one side brings down mortality rates from the disease, while on the other side, it reflects better accessibility to healthcare and diagnosis in an attempt to reduce mortality rates from such diseases across different geographies(RSNA, 2024).

## Visualizing the Distribution of Data via Histograms
Histograms are powerful tools for visualizing the distribution of numerical data, making them an essential part of exploratory data analysis. They provide a graphical representation that allows us to see the underlying frequency distribution of our data set.

Definition and Structure of Histograms: A histogram is constructed by dividing the range of the data into a series of intervals, known as bins. Each bin represents a specific range of values, and the height of each bar indicates the count of observations that fall within that range. This process is akin to creating a bar graph, but instead of showing categorical data, histograms display continuous numerical variables.

How Histograms Work:

Binning the Data: To create a histogram, you start by selecting appropriate bin widths. The choice of bin size can significantly impact the visualization. Too few bins may oversimplify the data, while too many bins may overcomplicate it, leading to noise.

Counting Observations: Once the data is binned, the next step is to count the number of data points that fall into each bin. This count is represented by the height of the bar for that bin.

Displaying Frequencies: The histogram visually summarizes the distribution of the data, allowing you to quickly assess the frequency of different ranges of values.

Interpreting Histograms:

Shape of Distribution: By examining the shape of the histogram, you can gain insights into the distribution characteristics of your data.
Gaussian Distribution: If the histogram resembles a bell curve, it suggests that the data follows a normal distribution, which is symmetric around the mean.
Skewness: If one tail of the histogram is longer or fatter than the other, the distribution may be skewed. A right (or positive) skew means that the tail on the right side is longer, while a left (or negative) skew indicates the opposite.
Exponential Distribution: A histogram showing a rapid decline in frequency as you move away from the peak may indicate an exponential distribution.
Identifying Outliers: Histograms can also help identify outliers—data points that lie far outside the range of the majority of the data. These outliers can significantly affect statistical analyses, so recognizing them is crucial.
Practical Applications of Histograms:

Data Analysis: Histograms are widely used in various fields such as finance, biology, and engineering for analyzing data distributions.
Quality Control: In manufacturing and quality control, histograms help monitor product characteristics and determine if they meet specified limits.
Statistical Modeling: Understanding the distribution of your data is vital for selecting appropriate statistical methods and models.

## Data Pre-processing
Data preprocessing is an essential step in the data analysis workflow, playing a pivotal role in the success of any machine learning project. Effective preprocessing ensures that the data is in the best possible shape to reveal the underlying structure of the problem at hand, thereby facilitating the machine learning algorithms' ability to learn from the data effectively. Properly prepared data can lead to improved model performance, accuracy, and interpretability.

The data preprocessing process typically involves several key activities:

Assigning Numerical Values to Categorical Data: Many machine learning algorithms require numerical input, so it is necessary to convert categorical variables into a format that can be easily understood by these algorithms. This may involve techniques such as one-hot encoding, where each category is transformed into a binary vector, or label encoding, where each category is assigned a unique integer value. ## Classification with Cross-Validation
splitting the data into training and test sets is essential for preventing overfitting. This approach ensures that the model can generalize effectively to real-world, previously unseen data. Cross-validation takes this concept a step further by introducing a systematic method for validating model performance through multiple iterations. Instead of relying on a single train/test split, cross-validation involves dividing the dataset into several equally sized subsets, known as "folds."

The training process begins by utilizing all but one of these folds for training, with the excluded fold serving as the holdout sample. After training the model on the selected folds, the next step is to evaluate its performance using this holdout sample. Once the evaluation is complete, the holdout sample is reintegrated with the other folds, and a different fold is designated as the new holdout sample.

This process of training the model and validating it on the holdout sample is repeated for each fold in the dataset, ensuring that every subset has the opportunity to act as the test or holdout sample at least once. The expected performance of the classifier, referred to as the cross-validation error, is calculated by averaging the error rates obtained from each holdout sample.

To illustrate this process, one can start with a standard train/test split, which serves as the baseline, and then proceed to compute the cross-validation error. This methodology not only provides a more reliable estimate of the model’s predictive performance but also helps in identifying potential issues related to overfitting by leveraging the entire dataset for training and validation in a structured manner.

Model Accuracy: Receiver Operating Characteristic (ROC) Curve
In the realms of statistical modeling and machine learning, one of the most widely used performance metrics for evaluating model accuracy in binary classification tasks is the Area Under the Curve (AUC). This metric provides a comprehensive assessment of the model's ability to distinguish between the two classes, typically labeled as positive and negative.

To grasp the information conveyed by the ROC curve, it is helpful to refer to the confusion matrix. This matrix serves as a two-dimensional table that displays the performance of the classifier model, with one axis representing the model's predictions (vertical) and the other axis representing the actual ground truth (horizontal). Each axis can take on two values, resulting in four possible outcomes:

Model says "+"	Model says "-"
True Positive	False Negative
Actual: "+"	
False Positive	True Negative
Actual: "-"	


In constructing the ROC curve, we plot the True Positive Rate (TPR) on the Y-axis against the False Positive Rate (FPR) on the X-axis. The values for true positives, false negatives, false positives, and true negatives are derived from the confusion matrix, and the rates are defined as follows:

True Positive Rate (TPR), also known as sensitivity:'
True positive rate (or sensitivity)}: tpr = tp / (tp + fn)
This metric reflects the proportion of actual positive cases that were correctly identified by the model.


False positive rate: fpr = fp / (fp + tn)
This rate measures the proportion of actual negative cases that were incorrectly classified as positive by the model.

True negative rate (or specificity): tnr = tn / (fp + tn)
This metric captures the proportion of actual negative cases that were correctly identified.

In these definitions, the denominator represents the row margins of the confusion matrix. The True Positive Rate (TPR) can be interpreted as the probability that the model predicts a positive outcome ("+") when the actual value is indeed positive. However, it is important to note that this does not provide insight into how likely the model is to be correct when it predicts a positive outcome. In other words, the TPR only describes the model's sensitivity without addressing the conditional probability of a true positive result based on the model's prediction of a positive class.

The ROC curve serves as a powerful tool for visualizing and analyzing the trade-offs between TPR and FPR at various threshold settings, ultimately enabling practitioners to select an appropriate threshold for classification based on the specific context and objectives of their analysis. ### Optimizing the SVM Classifier
Machine learning models, including Support Vector Machines (SVMs), are built with a variety of parameters that dictate their behavior and performance on specific tasks. These parameters are crucial because they can significantly influence the model's accuracy, generalization ability, and overall effectiveness in classifying data. The process of adjusting these parameters to find the best configuration for a particular problem is known as hyperparameter tuning.

SVMs, in particular, can have several parameters that need to be optimized. Some of the key parameters include:

Regularization parameter (C): This parameter controls the trade-off between maximizing the margin and minimizing the classification error. A smaller value of C allows for a larger margin but may lead to misclassification of training points, while a larger C value strives to classify all training points correctly, potentially leading to a smaller margin.

Kernel type: SVMs can utilize different kernel functions to transform the input data into higher dimensions, making it easier to find a hyperplane that separates the classes. Common kernel types include linear, polynomial, and radial basis function (RBF) kernels. The choice of kernel can significantly impact the performance of the classifier.

Kernel-specific parameters: Depending on the chosen kernel, there may be additional parameters to tune. For example, the RBF kernel requires the tuning of the gamma parameter, which influences the shape of the decision boundary. A high gamma value can lead to overfitting, while a low value may cause underfitting.

Given the potential complexity of these parameters, optimizing an SVM model can be treated as a search problem where we explore the parameter space to find the optimal combination. This optimization process can involve methods such as grid search or random search, which systematically evaluate different parameter combinations.

In this part of our analysis, we will utilize the powerful scikit-learn library to perform hyperparameter tuning for our SVM classification model. Scikit-learn provides convenient tools for implementing techniques like GridSearchCV, which automates the search for the best hyperparameters using cross-validation. By specifying a range of values for each parameter of interest, we can evaluate the performance of the model across different configurations.

The objective of this optimization process is to improve the model's performance metrics—such as accuracy, precision, recall, and the F1-score—by identifying the most effective parameter settings. Once we have tuned the SVM classifier and identified the optimal parameters, we can expect improved classification performance, leading to more accurate predictions when the model is applied to unseen data.

Ultimately, optimizing the SVM classifier not only enhances the model's ability to generalize well on new data but also deepens our understanding of how each parameter influences the overall behavior of the model, paving the way for more informed decisions in future modeling tasks.

```python
# Spot-Check Algorithms: Define a list of models to evaluate
models = []  # Initialize an empty list to store model names and instances
models.append(('LR', LogisticRegression()))  # Logistic Regression
models.append(('LDA', LinearDiscriminantAnalysis()))  # Linear Discriminant Analysis
models.append(('KNN', KNeighborsClassifier()))  # K-Nearest Neighbors
models.append(('CART', DecisionTreeClassifier()))  # Classification and Regression Trees
models.append(('NB', GaussianNB()))  # Naive Bayes Classifier
models.append(('SVM', SVC()))  # Support Vector Machine

# Test options and evaluation metric
num_folds = 10  # Define the number of folds for cross-validation
num_instances = len(X_train)  # Get the number of instances in the training data
seed = 7  # Set a random seed for reproducibility
scoring = 'accuracy'  # Define the scoring metric to evaluate models

# Initialize lists to store results and model names
results = []  # List to hold the accuracy results for each model
names = []  # List to hold the names of each model

# Evaluate each model using cross-validation
for name, model in models:  # Iterate through each model in the list
    kfold = KFold(n_splits=num_folds, random_state=seed, shuffle=True)  # Set up K-Fold cross-validation
    cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring=scoring)  # Perform cross-validation
    results.append(cv_results)  # Append the results to the results list
    names.append(name)  # Append the model name to the names list
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())  # Create a message with mean and std dev of accuracy
    print(msg)  # Print the model name and its cross-validation results

# Final message indicating the completion of the cross-validation process
print('\n➔ 10-Fold cross-validation accuracy score for the training data for six classifiers')
```

Importance of Distribution Analysis
Understanding Variability: Analyzing the distribution of accuracy scores reveals how stable and reliable the model's performance is. A model that performs well on average but exhibits high variability may not be suitable for deployment, as it could yield inconsistent results in practice.

Identifying Outliers: A box-and-whisker plot provides a visual summary that highlights outliers and the overall range of accuracy scores. This can help identify specific cases where the model may struggle, allowing for targeted improvements or adjustments.

Comparing Model Performance: When multiple models are being evaluated, visualizing their performance distributions can aid in making informed decisions about which models to pursue further. It allows for a clear comparison of stability and performance consistency among different algorithms.

Visual Representation with Box and Whisker Plots
Box and whisker plots are particularly useful for this analysis as they display key statistical measures of the accuracy distributions:

Median: The line inside the box represents the median accuracy, providing a measure of central tendency.
Interquartile Range (IQR): The box itself spans the first and third quartiles (Q1 and Q3), indicating the middle 50% of accuracy scores, which gives an idea of the model's consistency.
Whiskers: The lines extending from the box (whiskers) indicate the range of accuracy scores outside the IQR, helping to visualize the full extent of model performance.
Outliers: Points that fall outside the whiskers are considered outliers and may warrant further investigation.
By employing box and whisker plots, we can gain valuable insights into the performance of Logistic Regression and LDA, guiding our decisions on whether to explore these models further or make adjustments to enhance their predictive capabilities. Overall, a more detailed analysis of the accuracy distributions will lead to a better understanding of model reliability and robustness in various scenarios.

This observation is encouraging, as it suggests that most classifiers exhibit low variance in their performance, leading to reliable predictions. The satisfactory results obtained from the SVM model further reinforce its effectiveness in this context.

However, the varied distribution of the input attributes may have an impact on the SVM's accuracy. SVMs are sensitive to the scale of the data, and differences in feature distribution can affect how well the model can separate classes. To address this potential issue, we will conduct a follow-up analysis using a standardized version of the training dataset. This standardization process will ensure that all features have a mean of zero and a standard deviation of one, allowing for a more equitable evaluation of the model’s performance and potentially improving the accuracy of the SVM classifier. By doing so, we aim to gain deeper insights into the effects of attribute distribution on model accuracy.

```python
# Evaluation of Algorithms
# Standardizing Data

# Standardize the dataset

# Initialize an empty list to hold the pipelines
pipelines = []

# Create a pipeline for Logistic Regression with standard scaling
pipelines.append(('ScaledLR', Pipeline([
    ('Scaler', StandardScaler()),  # Step to scale features
    ('LR', LogisticRegression())   # Logistic Regression model
])))

# Create a pipeline for Linear Discriminant Analysis with standard scaling
pipelines.append(('ScaledLDA', Pipeline([
    ('Scaler', StandardScaler()),  # Step to scale features
    ('LDA', LinearDiscriminantAnalysis())  # LDA model
])))

# Create a pipeline for K-Nearest Neighbors with standard scaling
pipelines.append(('ScaledKNN', Pipeline([
    ('Scaler', StandardScaler()),  # Step to scale features
    ('KNN', KNeighborsClassifier())  # KNN model
])))

# Create a pipeline for Decision Tree Classifier with standard scaling
pipelines.append(('ScaledCART', Pipeline([
    ('Scaler', StandardScaler()),  # Step to scale features
    ('CART', DecisionTreeClassifier())  # CART model
])))

# Create a pipeline for Gaussian Naive Bayes with standard scaling
pipelines.append(('ScaledNB', Pipeline([
    ('Scaler', StandardScaler()),  # Step to scale features
    ('NB', GaussianNB())  # Naive Bayes model
])))

# Create a pipeline for Support Vector Machine with standard scaling
pipelines.append(('ScaledSVM', Pipeline([
    ('Scaler', StandardScaler()),  # Step to scale features
    ('SVM', SVC())  # SVM model
])))

# Initialize lists to store results and names of models
results = []
names = []

# Evaluate each pipeline using cross-validation
for name, model in pipelines:
    # Set up K-Fold cross-validation
    kfold = KFold(n_splits=num_folds, random_state=seed, shuffle=True)
    
    # Perform cross-validation and calculate scores
    cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring=scoring)
    
    # Store the results and names
    results.append(cv_results)
    names.append(name)
    
    # Print the mean and standard deviation of cross-validation scores
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

```python
# Compare Algorithms

# Create a new figure for the comparison plot
fig = plt.figure()

# Set the title for the entire figure
fig.suptitle('Scaled Algorithm Comparison')

# Add a subplot to the figure (1 row, 1 column, first subplot)
ax = fig.add_subplot(111)

# Create a boxplot for the results of each algorithm
plt.boxplot(results)

# Set the x-tick labels to the names of the algorithms
ax.set_xticklabels(names)

# Display the plot
plt.show()
```

```python
#  Tuning algorithm
# Tuning hyper-parameters - SVC estimator 

# Make a pipeline for the Support Vector Classifier that includes scaling and PCA
pipe_svc = Pipeline([
    ('scl', StandardScaler()),  # Step 1: Scale the features
    ('pca', PCA(n_components=2)),  # Step 2: Apply PCA for dimensionality reduction
    ('clf', SVC(probability=True, verbose=False))  # Step 3: SVC classifier
])

# Fit the pipeline to the training data
pipe_svc.fit(X_train, y_train)

# Print a message indicating the pipeline has been fitted (commented out)
# print('➝ Fitted Pipeline to training Data')

# Perform cross-validation to assess model training accuracy
scores = cross_val_score(estimator=pipe_svc, X=X_train, y=y_train, cv=10, n_jobs=1, verbose=0)
print('➔ Model Training Accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))

# Define a range of values for the hyperparameter 'C' to tune
param_range = [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0]

# Define the parameter grid for GridSearchCV
param_grid = [
    {'clf__C': param_range, 'clf__kernel': ['linear']},  # For linear kernel
    {'clf__C': param_range, 'clf__gamma': param_range, 'clf__kernel': ['rbf']}  # For RBF kernel
]

# Set up GridSearchCV to find the best hyperparameters
gs = GridSearchCV(estimator=pipe_svc,
                  param_grid=param_grid,
                  scoring='accuracy',  # Use accuracy as the scoring metric
                  cv=10,  # 10-fold cross-validation
                  n_jobs=1)  # Number of jobs to run in parallel

# Fit the grid search to the training data
gs = gs.fit(X_train, y_train)

# Print the best score obtained from the tuned parameters
print('➔ Tuned Parameters Best Score: ', gs.best_score_)

# Print the best parameters found by the grid search
print('➔ Best Parameters: \n', gs.best_params_)
```

## Tuning the Hyperparameters: k-NN Hyperparameters
In a standard implementation of the k-Nearest Neighbors (k-NN) algorithm, two primary hyperparameters play a critical role in determining the model's performance:
The Number of Neighbors (k): This hyperparameter defines how many neighboring points are considered when making a prediction for a new data point. The choice of 
𝑘 can significantly influence the classifier's accuracy. A small k can lead to a model that is sensitive to noise in the training data (high variance), while a large k may oversimplify the model and ignore potentially valuable patterns (high bias).

The Distance Metric/Similarity Function: This hyperparameter determines how the distance between data points is calculated. Common distance metrics include Euclidean distance, Manhattan distance, and Minkowski distance, among others. The choice of distance metric can greatly affect the k-NN classifier's ability to correctly identify neighbors, thus influencing its overall accuracy.

Implementing Hyperparameter Tuning with Grid Search
To optimize the performance of the k-NN classifier, we can use the GridSearchCV utility from scikit-learn. This tool enables systematic tuning of hyperparameters by evaluating a range of values through cross-validation. Here’s how we can set up the tuning process:

10-Fold Cross-Validation: We will employ 10-fold cross-validation to assess the performance of different hyperparameter combinations. This means the dataset will be split into ten subsets, and the model will be trained and validated ten times, each time using a different subset as the validation set and the remaining subsets for training. This approach helps ensure that the model's performance is robust and not overly reliant on a single train/test split.

Parameter Grid Specification: To conduct the grid search, we need to create a parameter grid that specifies the values for k and the distance metric to be evaluated. For the 𝑛 neighbors parameter, we can set it to take values from 1 to 30. Instead of providing just a single list of values, we will specify a range for 𝑛 neighbors to allow GridSearchCV to test all values within that range.

Parallel Computation: To enhance efficiency, we can set the n_jobs parameter to -1. This setting allows the grid search process to utilize all available CPU cores for parallel computation, significantly reducing the time required to evaluate all combinations of hyperparameters.

```python
from sklearn.neighbors import KNeighborsClassifier as KNN  # Import KNN classifier

# Create a pipeline for KNN that includes scaling and PCA
pipe_knn = Pipeline([
    ('scl', StandardScaler()),  # Step 1: Scale the features
    ('pca', PCA(n_components=2)),  # Step 2: Apply PCA for dimensionality reduction
    ('clf', KNeighborsClassifier())  # Step 3: KNN classifier
])

# Fit the pipeline to the training data
pipe_knn.fit(X_train, y_train)

# Perform cross-validation to assess model training accuracy
scores = cross_val_score(estimator=pipe_knn, 
                         X=X_train, 
                         y=y_train, 
                         cv=10,  # Use 10-fold cross-validation
                         n_jobs=1)  # Number of jobs to run in parallel
print('➝ Model Training Accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))

# Define a range of values for the hyperparameter 'n_neighbors' to tune
param_range = range(1, 31)  # Test values from 1 to 30 for n_neighbors

# Define the parameter grid for GridSearchCV
param_grid = [{'clf__n_neighbors': param_range}]  # Specify the hyperparameter to tune

# Instantiate the grid search with the KNN pipeline
gs = GridSearchCV(estimator=pipe_knn, 
                  param_grid=param_grid, 
                  cv=10,  # 10-fold cross-validation
                  scoring='accuracy')  # Use accuracy as the scoring metric

# Fit the grid search to the training data
gs = gs.fit(X_train, y_train)

# Print the best score obtained from the tuned parameters
print('➔ Tuned Parameters Best Score: ', gs.best_score_)

# Print the best parameters found by the grid search
print('➔ Best Parameters: \n', gs.best_params_)
```

```python
# Finalize Model 
# Use the best parameters found from the grid search
clf_svc = gs.best_estimator_  # Retrieve the best estimator from grid search

# Fit the final model to the training data
clf_svc.fit(X_train, y_train)

# Perform cross-validation to get final scores for the trained model
scores = cross_val_score(estimator=clf_svc,
                         X=X_train,
                         y=y_train,
                         cv=10,  # Use 10-fold cross-validation
                         n_jobs=1)  # Number of jobs to run in parallel

# Print the final training accuracy of the model
print('➔ Final Model Training Accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))

# Evaluate the model on the test set and print the accuracy
print('➜ Final Accuracy on Test set: %.5f' % clf_svc.score(X_test, y_test))
```

```python
# Fit the finalized Support Vector Classifier model to the training data
clf_svc.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf_svc.predict(X_test)

# Calculate and print the accuracy score of the model on the test set
print(accuracy_score(y_test, y_pred))

# Generate and print the confusion matrix to evaluate the model's performance
print(confusion_matrix(y_test, y_pred))

# Generate and print a classification report including precision, recall, and F1-score
print(classification_report(y_test, y_pred))
```
